```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

xticks = [f'DS_{i}' for i in range(1, 21)]
```

### Plot misclassified sequence numbers of Tiara and Whokaryote

```python
tiara = pd.read_csv('/home/tianqi/project/DeepMicrobeFinder/perf_summary/misclassified_tiara.csv')
tiara.index = xticks

whokaryote = pd.read_csv('/home/tianqi/project/DeepMicrobeFinder/perf_summary/misclassified_whokaryote.csv')
whokaryote.index = xticks

fig, ax = plt.subplots(1, 2, figsize=(12, 6))
tiara.plot.bar(stacked=True, ax=ax[0])
ax[0].set_title('Tiara')
ax[0].set_ylabel('Number of misclassified samples')
ax[0].set_ylim(0, 100)
ax[0].tick_params(axis='x', labelrotation=30)


whokaryote.plot.bar(stacked=True, ax=ax[1])
ax[1].set_title('Whokaryote')
ax[1].set_ylim(0, 500)
ax[1].tick_params(axis='x', labelrotation=30)

fig.tight_layout()
fig.supxlabel('Dataset')
fig.savefig('/home/tianqi/project/DeepMicrobeFinder/plot_figures/plot_acc_f1/analysis_results/mistake_euk.pdf', bbox_inches='tight')
```

### Plot misclassfied plasmid sequence numbers of Plasflow and PPR-Meta

```python
plasflow = pd.read_csv('/home/tianqi/project/DeepMicrobeFinder/perf_summary/misclassified_plasflow.csv')
plasflow.index = xticks

ppr = pd.read_csv('/home/tianqi/project/DeepMicrobeFinder/perf_summary/misclassified_ppr.csv')
ppr.index = xticks

fig, ax = plt.subplots(1, 2, figsize=(12, 6))
plasflow.plot.bar(stacked=True, ax=ax[0])
ax[0].set_title('Plasflow')
ax[0].set_ylabel('Number of misclassified samples')
ax[0].set_ylim(0, 400)
ax[0].tick_params(axis='x', labelrotation=30)

ppr.plot.bar(stacked=True, ax=ax[1])
ax[1].set_title('PPR')
ax[1].set_ylim(0, 400)
ax[1].tick_params(axis='x', labelrotation=30)

fig.tight_layout()
fig.supxlabel('Dataset')
fig.savefig('/home/tianqi/project/DeepMicrobeFinder/plot_figures/plot_acc_f1/analysis_results/mistake_pro.pdf', bbox_inches='tight')

```
