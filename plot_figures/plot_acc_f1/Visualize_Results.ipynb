### Order

```python
def getFirstNum(name):
    values = name.split("_")[:-1]
    return int(values[0])
```

```python
print(getFirstNum("357_71_71_416_83_test.txt"))
```

```python
def sortByRatio(names):
    return sorted(list(names), key= lambda x: int(x.split("_")[0]) + int(x.split("_")[1]))        
```

```python
ordered_list = sortByRatio(df_def_euk.filename)
```

```python
def reorder_df(df):
    df1 = df.set_index('filename')
    df2 = df1.reindex(ordered_list)
    return df2
```

```python
# t-test 
from scipy import stats
```

```python
def get_t_stats(df1, df2):
    t_f1_score = stats.ttest_ind(list(df1.f1_score), list(df2.f1_score), equal_var=False)
    t_acc = stats.ttest_ind(list(df1.accuracy), list(df2.accuracy), equal_var=False)
    print("t-test acc: {}".format(t_acc))
    print("t-test f1: {}".format(t_f1_score))
```

```python
get_t_stats(df_def_euk1, df_eukrep_1)
```

```python
get_t_stats(df_def_plasmid_1, df_plasflow_1)
```
